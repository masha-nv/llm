```python
from ddgs import DDGS
from fastcore.all import *

def search_images(keywords, max_images=200):
  return L(DDGS().images(keywords=keywords,max_results=max_images)).itemgot('image')
```

```python
urls = search_images(keywords='bird', max_images=1)
```

```python
from fastdownload import download_url
dest = 'bird.jpg'
download_url(url=urls[0], dest=dest)
```

```python
from fastai.vision.all import *
img = Image.open(dest)
img.to_thumb(256, 256)
```

```python
urls = search_images(keywords='national park', max_images=1)
url = urls[0]
dest = 'park.jpg'
download_url(url=url, dest=dest)
img = Image.open(dest)
img.to_thumb(256, 256)
```

```python
searches = 'mountains', 'bird', 'cat', 'kitten'
path = Path('ddgs_images')
for o in searches:
  dest = (path/o)
  dest.mkdir(exist_ok=True, parents=True)
  download_images(dest, urls=search_images(keywords=o, max_images=20))
  time.sleep(5) # Add a longer delay
  resize_images(dest, max_size=400, dest=dest)
```

```python
failed = verify_images(get_image_files(path=path))
failed.map(Path.unlink) # Explicitly remove broken images
len(failed)
```

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]).dataloaders(path, bs=10)
```

```python
dls.show_batch(max_n=6)
```

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
```

```python
learn.fine_tune(30)
```

```python
kitten = search_images(keywords='kitten', max_images=1)
kitten[0]
download_url(url=kitten[0], dest='kitten.jpg')
```

```python
img = Image.open('kitten.jpg')
img.to_thumb(200, 200)
```
